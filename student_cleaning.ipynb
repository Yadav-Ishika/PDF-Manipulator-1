### Let's Start with `head()`

- `head()` method returns first 5 rows of the DataFrame, indexed `0` to `4`

```python
df.head()
```

- `head(n)` method returns first n rows of the DataFrame, indexed `0` to `n-1`

```python
df.head(3)
```

- `tail()` method returns last 5 rows of the DataFrame, indexed `l-5` to `l` where `l` represents *the* number of rows

```python
df.tail()
```

- `tail(n)` method returns last n rows of the DataFrame, indexed `l-n` to `l` where l represents the number of rows

```python
df.tail(3)
```

- `shape` attribute returns dimensions of the DataFrame in order
(rows, column)

```python
df.shape
```

- `info()` method returns the information of DataFrame which includes details like `column names`, `column dtypes` & `number of columns`

```python
df.info()
```

- `describe()` method returns following information of DataFrame

```python
df.describe()
```

- `describe(include='requirements')` method returns following information of DataFrame

```python
df.describe(include='all')
```
